```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = 6
pos = np.arange(1,11)
xrange = np.linspace(0,10, 1000)
```

```python
pos_encoding = []
for i in range(d):
    if i%2 == 0:
        pos_encoding.append(np.sin(pos/10000**(i/d)))
    else:
        pos_encoding.append(np.cos(pos/10000**((i-1)/d)))
```

```python
sinusoidal = []
for i in range(d):
    if i%2 == 0:
        sinusoidal.append(np.sin(xrange/10000**(i/d)))
    else:
        sinusoidal.append(np.cos(xrange/10000**((i-1)/d)))
```

```python
fig, axes = plt.subplots(6,1, figsize = (10,6), sharex = True)
for j in range(d):
    axes[j].plot(xrange, sinusoidal[j], color = '#D62828', lw = 2)
    axes[j].scatter(pos, pos_encoding[j], color = '#FCBF49', marker = 'o', s = 36)
    axes[j].spines['top'].set_visible(False)
    axes[j].spines['right'].set_visible(False)
    axes[j].set_ylabel(f"2i = {j}")
    if j == 5:
        axes[j].set_xlabel('position')
        
fig.suptitle('Positional Encoding', fontweight="bold")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(-3, 3, 100)
sigmoid = 1/(1 + np.exp(-z))
tanh = (np.exp(z)-np.exp(-z))/(np.exp(z)+np.exp(-z))
linear = z
ReLU = np.where(z>=0,z, 0)
Leaky = np.where(z>=0.01*z,z,0.01*z)
```

```python
plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots(2,2, figsize = (10,10))
ax[0,0].plot(z, sigmoid, lw = 2)
ax[0,0].hlines(xmin = -3, xmax = 3, y = 0, color = "red", linestyle = "dashed")
ax[0,0].set_xlabel("z")
ax[0,0].set_ylabel("a")
ax[0,0].set_title("sigmoid", fontsize = 14)

ax[0,1].plot(z, tanh, lw = 2)
ax[0,1].hlines(xmin = -3, xmax = 3, y = 0, color = "red", linestyle = "dashed")
ax[0,1].set_xlabel("z")
ax[0,1].set_ylabel("a")
ax[0,1].set_title("tanh", fontsize = 14)

ax[1,0].plot(z, ReLU, lw = 2)
ax[1,0].hlines(xmin = -3, xmax = 3, y = 0, color = "red", linestyle = "dashed")
ax[1,0].set_xlabel("z")
ax[1,0].set_ylabel("a")
ax[1,0].set_title("ReLU", fontsize = 14)

ax[1,1].plot(z, Leaky, lw = 2)
ax[1,1].hlines(xmin = -3, xmax = 3, y = 0, color = "red", linestyle = "dashed")
ax[1,1].set_xlabel("z")
ax[1,1].set_ylabel("a")
ax[1,1].set_title("Leaky ReLU", fontsize = 14)

fig.tight_layout(pad=3.0)
plt.show()
```

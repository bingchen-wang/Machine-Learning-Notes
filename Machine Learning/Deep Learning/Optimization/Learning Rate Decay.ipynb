```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epochs = np.arange(100)
```

```python
decay_rate = 0.5
a0 = 0.1
a_denom = 1/(1 + decay_rate * epochs) * a0
a_exp = 0.95**(epochs)*a0
a_ksqrt = 1/np.sqrt(epochs+1)*a0
a_discrete = 0.8**(epochs//10)*a0
```

```python
plt.figure(figsize = (10,6))
plt.plot(epochs, a_denom, color = "#FF0000", label = r'$\alpha = \frac{1}{1 + 0.5(i-1)} \alpha_0$', lw = 2)
plt.plot(epochs, a_exp, color = "#00A08A",label = r'$\alpha = {0.95}^{i-1} \alpha_0$', lw = 2)
plt.plot(epochs, a_ksqrt, color = "#F2AD00",label = r'$\alpha = \frac{1}{\sqrt{i}}\alpha_0$', lw = 2)
plt.plot(epochs, a_discrete, color = "#5BBCD6",label = r'$\alpha = {0.8}^{\left\lfloor \frac{i-1}{10} \right\rfloor} \alpha_0$', lw = 2)
plt.legend(fontsize = 20)
plt.title('Different learning rate decay methods', fontsize = 16)
plt.xlabel(r'Epoch $i$', fontsize = 14)
plt.ylabel(r'Learning rate $\alpha$', fontsize = 14)
plt.show()
```

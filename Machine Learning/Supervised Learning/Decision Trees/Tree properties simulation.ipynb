```python
import numpy as np
import matplotlib.pyplot as plt
import math

n = np.arange(1,100)
n_1 = (1-1/n)**n
plt.figure(figsize = (8,6))
plt.plot(n, n_1, color = "darkblue", linestyle = "dotted", label = r'${(1-\frac{1}{n})}^n$')
plt.scatter(n, n_1, color = "skyblue", alpha = 0.75)
plt.hlines(xmin = 1, xmax = 100, y =1/3, color = "red", label = r'$1/3$')
plt.hlines(xmin = 1, xmax = 100, y =1/math.e, color = "purple",  label = r'$1/e$')
plt.legend(fontsize = 12, loc = 'lower right')
plt.xlabel('n', fontsize = 12)
plt.ylabel('y', fontsize = 12)
plt.title(r'Convergence of ${(1-\frac{1}{n})}^n$', fontsize = 16)
plt.savefig('convergence to 1 over e.png')
```

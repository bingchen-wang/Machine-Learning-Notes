```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
kmeans = KMeans(n_clusters = 3, init='k-means++', random_state = 0)
kmeans.fit(X)

fig, ax = plt.subplots(1,2, figsize = (10,5), subplot_kw=dict(projection='3d'))
ax[0].scatter(X[:,0], X[:,1],X[:,2], c = kmeans.labels_)
ax[0].set_xlabel("Sepal length")
ax[0].set_ylabel("Sepal width")
ax[0].set_zlabel("Petal length")
ax[0].set_title("3D plot with predicted labels (K-means Clustering)")

ax[1].scatter(X[:,0], X[:,1],X[:,2], c = y)
ax[1].set_xlabel("Sepal length")
ax[1].set_ylabel("Sepal width")
ax[1].set_zlabel("Petal length")
ax[1].set_title("3D plot with actual labels")

plt.show()

# For large datasets (say n_samples > 10k), try MiniBatchKMeans.
```
